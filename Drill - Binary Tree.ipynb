```python
import sys
    
def get_breadth_first_nodes(root):
    nodes = []
    stack = [root]
    while stack:
        cur_node = stack[0]
        stack = stack[1:]
        nodes.append(cur_node)
        for child in cur_node.get_children():
            stack.append(child)
    return nodes

def get_depth_first_nodes(root):
    nodes = []
    stack = [root]
    while stack:
        cur_node = stack[0]
        stack = stack[1:]
        nodes.append(cur_node)        
        for child in cur_node.get_rev_children():
            stack.insert(0, child)
    return nodes

########################################################################
class Node(object):
    def __init__(self, id_):
        self.id = id_
        self.children = []
        
    def __repr__(self):
        return "Node: [%s]" % self.id
    
    def add_child(self, node):
        self.children.append(node) 
    
    def get_children(self):
        return self.children         
    
    def get_rev_children(self):
        children = self.children[:]
        children.reverse()
        return children         

########################################################################
def println(text):
    sys.stdout.write(text + "\n")
    
def make_test_tree():
    a1 = Node("a1")
    b2 = Node("b2")      
    b3 = Node("b3")      
    c4 = Node("c4")      
    c5 = Node("c5")      
    c6 = Node("c6")  
    c7 = Node("c7") 
    d8 = Node("d8")
    d9 = Node("d9") 
    d10 = Node("d10") 
    d11 = Node("d11") 
    d12 = Node("d12") 
    d13 = Node("d13") 
    d14 = Node("d14") 
    d15 = Node("d15") 
    
    a1.add_child(b2) 
    a1.add_child(b3)
    
    b2.add_child(c4)
    b2.add_child(c5)
    
    b3.add_child(c6) 
    b3.add_child(c7) 
    
    c4.add_child(d8) 
    c4.add_child(d9) 
    
    c5.add_child(d10) 
    c5.add_child(d11)
    
    c6.add_child(d12) 
    c6.add_child(d13)
    
    c7.add_child(d14) 
    c7.add_child(d15)
    
    return a1                  

def test_breadth_first_nodes():
    root = make_test_tree()
    node_list = get_breadth_first_nodes(root)
    for node in node_list:
        println(str(node))

def test_depth_first_nodes():
    root = make_test_tree()
    node_list = get_depth_first_nodes(root)
    for node in node_list:
        println(str(node))

########################################################################
if __name__ == "__main__":
    test_breadth_first_nodes()
    println("")
    test_depth_first_nodes()
  
```
