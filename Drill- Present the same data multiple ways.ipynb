```python
ax = sns.violinplot(x="room_type", y="price", data=listings_temp)
```

- Pros: . It shows the distribution of quantitative data across several levels of one (or more) categorical variables such that those distributions can be compared
- cons: violins for relatively small samples might look misleadingly smooth.

```python
ax = sns.lvplot(x="room_type", y="price", data=listings_temp)
```

- Pros:This plot shows distributions of large datasets, they are fast to generate, directly interpretable in terms of the distribution of data, and easy to understand
- Cons: Helpful for large datasets

```python
ax = sns.pointplot(x="room_type", y="price", data=listings_temp)
```

- Pros: They are particularly adept at showing interactions: how the relationship between levels of one categorical variable changes across levels of a second categorical variable. The lines that join each point from the same hue level allow interactions to be judged by differences in slope, which is easier for the eyes than comparing the heights of several groups of points or bars.
- Cons: a point plot shows only the mean (or other estimator) value, but in many cases it may be more informative to show the distribution of values at each level of the categorical variables. In that case, other approaches such as a box or violin plot may be more appropriate.
