```python
df = pd.read_excel("/Users/saurabh/Documents/PythonData/table_8_offenses_known_to_law_enforcement_california_by_city_2013.xls", encoding = "ISO-8859-1", skiprows=4)
```

```python
df.fillna(df.mean(),inplace=True)
```

```python
df = df.drop(['Rape\n(revised\ndefinition)1'], axis=1)
```

```python
df.loc[df['Murder and\nnonnegligent\nmanslaughter']>0,'Murder']=1
df.loc[df['Murder and\nnonnegligent\nmanslaughter']==0,'Murder']=0
```

```python
df.head()
```

```python
from sklearn.model_selection import train_test_split
X = df.drop(['Murder','City'],1)
y = df['Murder']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
#scaled the data
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_train_scaled = pd.DataFrame(scaler.fit_transform(X_train), columns=X_train.columns)
```

```python
X_test_scaled = pd.DataFrame(scaler.transform(X_test), columns=X_test.columns)
```

### Vanilla logistic regression

```python
lr = LogisticRegression(C=1e9)

# Fit the model.
fit = lr.fit(X_train_scaled, y_train)


# Display.
print('Coefficients')
print(fit.coef_)
print(fit.intercept_)
pred_y_sklearn = lr.predict(X_train_scaled)

#print('\n Accuracy by admission status')
#print(pd.crosstab(pred_y_sklearn, training_scores_encoded))

print('\n Percentage accuracy')
print(lr.score(X_train_scaled, y_train))
```

```python
print(lr.score(X_test_scaled, y_test))
```

### Ridge logistic regression

```python
clf1 = LogisticRegression(penalty='l2')
clf1.fit(X_train_scaled, y_train)
print('\nAccuracy:')
print(clf1.score(X_train_scaled, y_train))
print(clf1.score(X_test_scaled, y_test))
```

```python
clf1.coef_
```

### Lasso logistic regression

```python
clf2 = LogisticRegression(penalty='l1')
clf2.fit(X_train_scaled, y_train)
print('\nAccuracy:')
print(clf2.score(X_train_scaled, y_train))
print(clf2.score(X_test_scaled, y_test))
```

```python
clf2.coef_
```

The above analysis shows that lasso regression worked better than ridge also losgistic regression worked better than any other regression.
There are some stregths as well as limitations of regression as modeling approach
- Lasso can be costly some times as it removes columns which may leads to loss of information in some cases while in other cases that may be useful
- Ridge does not remove columns so that can be useful in some cases as you will not be loosing much of data but may be not good if that was least important data
